```python
import pandas as pd
from pandas import DataFrame as info
from matplotlib import style, pyplot as plot
from datetime import datetime

info = pd.read_csv('covid_19_india.csv', index_col=1, parse_dates=True, dayfirst=True)

start_date = '2020-06-01'
end_date   = '2020-06-30'

info = info.loc[start_date : end_date]

dFrame = info.loc[(info['State/UnionTerritory'] == 'Maharashtra')]

info = pd.DataFrame(dFrame)

info= info.reset_index()

info.get(['Sno', 'Date', 'State/UnionTerritory', 'Cured', 'Deaths', 'Confirmed'])

for ind in info.index: 
    #Extract data row by row
    info['Date'][ind], info['State/UnionTerritory'][ind], info['Cured'][ind], info['Deaths'][ind], info['Confirmed'][ind]
info



dateList = [i.strftime("%d") for i in info['Date'].tolist()]


plot.plot(dateList,info['Confirmed'].tolist(),'b', label='Confirmed In Maharashtra', linewidth =4)
plot.plot(dateList,info['Cured'].tolist(),'g', label='Cured In Maharashtra', linewidth =4)
plot.plot(dateList,info['Deaths'].tolist(),'r', label='Deaths In Maharashtra', linewidth =4)

plot.title('COVID-19 RECORD DATA ')
plot.xlabel('Days Of Month(June)')
plot.ylabel('Number Of People')
plot.legend()

plot.grid(True, color='y')
plot.legend(loc='best')
plot.savefig('Maharashtra.jpg')
plot.show()
```
